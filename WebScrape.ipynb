```python
from bs4 import BeautifulSoup
import requests

mountains = {}
response = requests.get("https://mountain-forecast.com/countries/Philippines?top100=yes")
response.raise_for_status()  # Ensure we notice bad responses
soup = BeautifulSoup(response.text, 'html.parser')
items = soup.find_all('li', class_='b-list-table__item')


for item in items:
        name = item.find('span', class_='b-list-table__item-name').get_text(strip=True).replace(" ", "-").replace( "(", "" ).replace( ")", "" )
        height = item.find('span', class_='b-list-table__item-height').get_text(strip=True).replace("m", "").strip()
        mountains[name] = height

print(mountains)
```

```python
url_list = []
for name, height in mountains.items():
    url_list.append(f"https://mountain-forecast.com/peaks/{name}/forecasts/{height}")
```
